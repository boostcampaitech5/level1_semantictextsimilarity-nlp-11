```python
import pandas as pd
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)
```

```python
models = ["bongsoo", "kykim", "monologg", "snunlp", "xlm"]
```

```python
model_output_dict = dict()
total_length = 1100
esnb = np.zeros((total_length, len(models)))

for model in models:
    model_output_dict[model] = pd.read_csv(f'./esnb_ingredients/output_{model}.csv')


for i in range(total_length):
    for j, model in enumerate(models):
        esnb[i, j] = model_output_dict[model]["target"][i]
```

```python
esnb
```

```python
esnb_result = np.zeros((total_length, 2))
esnb_std_mean_list = np.zeros((total_length, 2))
esnb_result
```

```python
for i in range(total_length):
    # esnb_result[i] = sum(esnb[i]) - max(esnb[i]) 
    esnb_std_mean_list[i, 0] = np.std(esnb[i])
    esnb_std_mean_list[i, 1] = np.mean(esnb[i])
```

```python
for i in range(total_length):
    for j in range(len(models)):
        if abs(esnb[i, j] - esnb_std_mean_list[i, 1]) / esnb_std_mean_list[i, 0] < 1.04:
            esnb_result[i][0] += esnb[i, j]
            esnb_result[i][1] += 1
```

```python
esnb_std_mean_list
```

```python
esnb_result
```

```python
esnb_result_final = np.zeros((total_length))
for i in range(total_length):
    esnb_result_final[i] = esnb_result[i][0] / esnb_result[i][1]
```

```python
esnb_result_final
```

```python
df = pd.DataFrame({"id": [f"boostcamp-sts-v1-test-{i:03}" for i in range(1100)],"target": esnb_result_final})
```
